Handling Outliers
An Outlier is a data-item/object that deviates significantly from the rest of the (so-called normal)objects. They can be caused by measurement or execution errors. The analysis for outlier detection is referred to as outlier mining. There are many ways to detect the outliers, and the removal process is the data frame same as removing a data item from the pandas dataframe

```python


sns.boxplot(x='sepal_width', data=iris)
```

```python
# In the above graph, the values above 4 and below 2 are acting as outliers.
```

Removing Outliers
For removing the outlier, one must follow the same process of removing an entry from the dataset using its exact position in the dataset because in all the above methods of detecting the outliers end result is the list of all those data items that satisfy the outlier definition according to the method used.

```python
# Importing
import sklearn
import pandas as pd
import seaborn as sns

# IQR
Q1 = np.percentile(iris['sepal_width'], 25,
				interpolation = 'midpoint')

Q3 = np.percentile(iris['sepal_width'], 75,
				interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", iris.shape)

# Upper bound
upper = np.where(iris['sepal_width'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(iris['sepal_width'] <= (Q1-1.5*IQR))

# Removing the Outliers
iris.drop(upper[0], inplace = True)
iris.drop(lower[0], inplace = True)

print("New Shape: ", iris.shape)

sns.boxplot(x='sepal_width', data=iris)
```

Build a Model

3. KNN Classification

```python
# Initially, the dataset is divided into features (X) and labels (y). The features include the measurements of the Iris flowers, and the labels represent the species. This separation is fundamental for supervised learning where the goal is to predict the labels based on the features.
# Splitting our datasets into its attributes and labels
# The X-variable containts the first four columns of the dataset(ie.Attributes ) while the y contains the labels.
X= iris.iloc[:,:-1].values
y= iris.iloc[:,4].values
```

```python
# To avoid over-fitting, we will dividide our dataset into training and test splits, which gives us a better idea as to how our algorithms perfromed during the testing phase.
# This was our algorithm is tested on un-seen data.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
# Feature Scaling
# Before making any actual predictions, it is always a good practice to scale the features so that all of the will be uniformly evaluated.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Training and Prediction
# The firs step is to import the KNeighborClassifier class from the sklearn.neighbor library.
# In the second line,this class is initialized with parameter ie. n_neighbors.
# This is basically the value for the K. K=5 is selected value for the KNN algorithm.
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
# Prediction:
# Predictions are made on the test data. This step is the culmination of the model training and is used to assess how well the model has learned from the training data.

y_pred = classifier.predict(X_test)
y_pred
```

```python
# Model Evaluation:
# For evaluating an algorithm, confusion matrix, precision, recall and f1 are most commonly used metrics.
# The confusion_matrix and classification_report methods of the sklearn_metrics can be used to calculate these metrics.

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#The provided data represents the performance of a classifier on the Iris dataset, with a confusion matrix and a classification report.
#The confusion matrix shows that all Iris-setosa and Iris-virginica instances were correctly classified, but one Iris-versicolor was misclassified as Iris-virginica.
#The classification report reveals high precision, recall, and F1-scores for each class, indicating a high overall accuracy of 97% for the model.
```
